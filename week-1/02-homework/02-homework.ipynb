### Exercise 2: Boolean Expressions
* Each scenario below describes some data to be manipulated by your code. Choose appropriate variable names (and values) and create Boolean expressions for these scenarios.
  

1. Given the number of hours worked in a week, determine whether the employee will receive any overtime pay.
  

```python
# Exercise 2 - Part 1
hours_worked = float(input("How many have you worked? "))
hourly_wage = 15 #Amount paid per hour
paycheck = 0 #Declare paycheck variable



if (hours_worked <= 40): #Excute if hours worked are less than 40
  paycheck = hourly_wage * hours_worked
  print(paycheck)
elif(hours_worked > 40): #Execute if hours worked greater than 40
  paycheck = hourly_wage * 40
  overtime_pay = (hours_worked - 40) * (hourly_wage * 1.5)
  paycheck = paycheck + overtime_pay
  print(paycheck)



```
